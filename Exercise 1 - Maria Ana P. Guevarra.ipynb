1. Load the Shakespeareâ€™s Hamlet from the NLTK (shakespeare-hamlet.txt). After loading the data, perform the following operations on the data. Submit the notebook that does the tasks described below. 

```python
import nltk, re 
nltk.download('book', quiet=True) 
 
hamlet = nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt') 
print(hamlet) 
```

1.a Words containing 2 vowels characters

```python
a = re.findall(r'\b[^\WaAeEiIoOuU]*[aAeEiIoOuU][^\WaAeEiIoOuU]*[aAeEiIoOuU][^\WaAeEiIoOuU]*\b', hamlet) 
for i in a: 
    print(i) 
```

1.b Words that contain a hyphen  

```python
a = re.findall(r'\b\w*[-]\w*\b', hamlet) 
for i in a: 
    print(i) 
```

1.c Words in their present progressive form (ends in â€“ing)

```python
a = re.findall(r'\b(\w+ing)\b', hamlet) 
for i in a: 
    print(i) 
```

1.d Words in their superlative form (ends in â€“est)

```python
a = re.findall(r'\b(\w+est)\b', hamlet) 
for i in a:
    print(i)
```

1.e Words at the end of the sentence (next to sentence ending punctuations)  

```python
hamlet = nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt') 
a = re.findall(r'(?<=\W)[A-Za-z-]+[.!?](?=\W)', hamlet) 
for i in a: 
    print(i) 
```

2. Tokenize the data using Penn Tree Bank Tokenization  

```python
ptb_tokenizer = nltk.tokenize.treebank.TreebankWordTokenizer() 
a = ptb_tokenizer.tokenize(hamlet) 
for i in a: 
    print(i) 
```

3. Apply case folding on the tokens  

```python
ptb_tokenizer = nltk.tokenize.treebank.TreebankWordTokenizer() 
a = ptb_tokenizer.tokenize(hamlet) 
casefolding = [w.lower() for w in a] 
for i in casefolding: 
   print(i) 
```

4. Lemmatize the case folded tokens using Word Net Lemmatizer  

```python
ptb_tokenizer = nltk.tokenize.treebank.TreebankWordTokenizer() 
a = ptb_tokenizer.tokenize(hamlet) 
casefolding = [w.lower() for w in a] 
lemmatizer = nltk.WordNetLemmatizer() 
WNL = [lemmatizer.lemmatize(w) for w in casefolding] 
for i in WNL: 
    print(i) 
```

5. From the lemmatized tokens, calculate the following metrics 

    a. Word type (Number of unique tokens) 

```python
ptb_tokenizer = nltk.tokenize.treebank.TreebankWordTokenizer() 
a = ptb_tokenizer.tokenize(hamlet) 
casefolding = [w.lower() for w in a] 
lemmatizer = nltk.WordNetLemmatizer() 
WNL = [lemmatizer.lemmatize(w) for w in casefolding] 
 
unique = [] 
count = 0 
for i in WNL: 
    if i not in unique: 
        unique.append(i) 
 
print("Number of Unique tokens = ",len(unique), "\n") 
 
for i in unique: 
    print(i) 
```
